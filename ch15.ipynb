##Card

```
class Card:
    suits=['spades','hearts','diamonds','clubs']

    values=[None,None,
            '2','3','4','5','6','7','8','9',
            '10','Jack','Queen','King','Ace']

    def __init__(self,v,s):
        self.value=v
        self.suit=s

    def __lt__(self,c2):    #Less Than:未満
        if self.value<c2.value:
            return True
        if self.value==c2.value:
            if self.suit<c2.suit:
                return True
            else:
                return False
        else:
            return False

    def __gt__(self,c2):    #Greater Than:より大きい
        if self.value>c2.value:
            return True
        if self.value==c2.value:
            if self.suit>c2.suit:
                return True
            else:
                return False
        else:
            return False

    def __repr__(self):
        v=self.values[self.value]+' of '+self.suits[self.suit]
        return v
```

```
card1=Card(10,1)
card2=Card(11,3)
print(card1)
print(card2)
print(card1<card2)
print(card1>card2)
```

###Deck

```
from random import shuffle

class Deck:
    def __init__(self):
        self.cards=[]
        for i in range(2,15):
            for j in range(4):
                self.cards.append(Card(i,j))
        shuffle(self.cards)

    def rm_card(self):
        if len(self.cards)==0:
            return
        else:
            return self.cards.pop()
```

```
deck=Deck()
for card in deck.cards:
    print(card)
print('\n'+str(deck.rm_card())+' is removed')
print('\n')
for card in deck.cards:
    print(card)
```

###Player

```
class Player:
    def __init__(self,name):
        self.wins=0
        self.name=name
```

###Game

```
class Game:
    def __init__(self):
        name1=input('プレイヤー1の名前: ')
        name2=input('プレイヤー2の名前: ')
        self.deck=Deck()
        self.p1=Player(name1)
        self.p2=Player(name2)

    def wins(self,winner):
        win_msg='このラウンドは{}が勝ちました'.format(winner)
        print(win_msg)

    def draw(self,p1_name,p1_card,p2_name,p2_card):
        draw_msg='{}は{}を、{}は{}を引きました'
        draw_msg=draw_msg.format(p1_name,p1_card,p2_name,p2_card)
        print(draw_msg)

    def play_game(self):
        cards=self.deck.cards
        print('戦争を始めます!')
        while len(cards) >= 2:
            msg='qで終了、それ以外のキーでPlay: '
            response=input(msg)
            if response=='q':
                break
            p1_card=self.deck.rm_card()
            p2_card=self.deck.rm_card()
            p1_name=self.p1.name
            p2_name=self.p2.name

            self.draw(p1_name,p1_card,p2_name,p2_card)
            if p1_card > p2_card:
                self.p1.wins+=1
                self.wins(self.p1.name)
            else:
                self.p2.wins+=1
                self.wins(self.p2.name)
            
        win=self.winner(self.p1,self.p2)
        print('ゲーム終了!{}の勝利です!'.format(win))

    def winner(self,p1,p2):
        if p1.wins > p2.wins:
            return p1.name
        if p1.wins < p2.wins:
            return p2.name
        return '引き分け!'
```

```
game=Game()
game.play_game()
```
