```
class Lion:
    def __init__(self,name):
        self.name=name

    def __repr__(self): #objectをprint関数に渡したときに呼ばれる__repr__メソッドをオーバーライド
        return self.name
```

```
class AlwaysPositive:
    def __init__(self,number):
        self.n=number

    def __add__(self,other):    #__add__メソッドをクラスに持たせれば、オブジェクト同士を足し算の演算子で計算できる
        return abs(self.n+other.n)
```

```
x=AlwaysPositive(-20)
y=AlwaysPositive(10)
print(x+y)
```

```
class Square:
    square_list=[]  #クラス変数
    def __init__(self,number):
        if len(self.square_list)<4:
            self.square_list.append(number)

    def __repr__(self):
        for i,number in enumerate(self.square_list):
            self.square_list[i]=str(number)
        return ' by '.join(self.square_list)
```

```
square=Square(29)
square=Square(32)
square=Square(22)
square=Square(12)

print(square)
```

```
class Bob:
    def __init__(self):
        pass

def same_object(object1,object2):
    if object1 is object2:
        return True
    else:
        return False
```

```
bob=Bob()
ken=Bob()

print(same_object(bob,ken))

bob=Bob()
ken=bob

print(same_object(bob,ken))
```
