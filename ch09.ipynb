###2．何か質問するプログラムを書こう。入力された回答をファイルに書き出そう。

```
import csv
import os

name=input('名前を入力してください')
age=input('年齢を入力してください')

if os.path.isfile(ROOT_PATH+'/ch09_2.csv') is False:
    with open(ROOT_PATH+'/ch09_2.csv','a') as f:
        w=csv.writer(f,delimiter=',')
        w.writerow(['name','age'])
        w.writerow([name,age])
else:
    with open(ROOT_PATH+'/ch09_2.csv','a') as f:
        w=csv.writer(f,delimiter=',')
        w.writerow([name,age])
```

###3．リストのリストに含まれている要素をCSVファイルに書き出そう。
データは次の通り:[["TopGun","RiskyBusiness","MinorityReport"],["Titanic","TheRevenant","Inception"],["TrainingDay","ManonFire","Flight"]]  
データの各要素はCSVの1行となり、その1行に含まれる各要素がカンマで区切られるように書き出そう。

```
import csv

data=[["TopGun","RiskyBusiness","MinorityReport"],["Titanic","TheRevenant","Inception"],["TrainingDay","ManonFire","Flight"]]
with open(ROOT_PATH+'/ch09_3.csv','w') as f:
    w=csv.writer(f,delimiter=',')
    for row in data:
        w.writerow(row)
```

###4．チャレンジ3を日本語で同じようにやってみよう。

```
import csv

data=[["トップガン","リスキービジネス","マイノリティレポート"],["タイタニック","レブナント","インセプション"],["トレーニングデイ","マノンファイア","フライト"]]
with open(ROOT_PATH+'/ch09_4.csv','w') as f:
    w=csv.writer(f,delimiter=',')
    for row in data:
        w.writerow(row)
```
